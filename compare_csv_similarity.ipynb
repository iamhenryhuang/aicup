```python
import pandas as pd

# 讀取兩個檔案
r1 = pd.read_csv('../result.csv')
r2 = pd.read_csv('../predict_results/result(0.120603)')

print("=== result.csv 統計 ===")
print(f"總數: {len(r1)}")
print(f"label=1: {r1['label'].sum()} ({r1['label'].mean():.2%})")
print(f"label=0: {(r1['label']==0).sum()}")

print("\n=== result(0.120603).csv 統計 ===")
print(f"總數: {len(r2)}")
print(f"label=1: {r2['label'].sum()} ({r2['label'].mean():.2%})")
print(f"label=0: {(r2['label']==0).sum()}")

# 合併比對
merged = r1.merge(r2, on='acct', suffixes=('_new', '_old'))
same = (merged['label_new'] == merged['label_old']).sum()
diff = (merged['label_new'] != merged['label_old']).sum()

print(f"\n=== 相似度分析 ===")
print(f"相同預測: {same} / {len(merged)} ({same/len(merged):.2%})")
print(f"不同預測: {diff} / {len(merged)} ({diff/len(merged):.2%})")

if diff > 0:
    print(f"\n差異詳情:")
    print(f"  新版預測1但舊版預測0: {((merged['label_new']==1) & (merged['label_old']==0)).sum()}")
    print(f"  新版預測0但舊版預測1: {((merged['label_new']==0) & (merged['label_old']==1)).sum()}")
```
